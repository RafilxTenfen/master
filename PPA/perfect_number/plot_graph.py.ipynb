```python
import csv
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
COL_SECONDS=0
COL_LIMIT=1
COL_THREADS=2
COL_SCHEDULE=3

data_list=[]

with open('data_cct.csv', 'r', newline='\n') as csv_file:
    header = True
    data = csv.reader(csv_file, delimiter=';')
    for row in data:
        # skip header         
        if header:
            header = False
            continue
            
#         print(row, type(row), row[COL_SECONDS], type(row[COL_SECONDS]))
        
        row[COL_SECONDS] = float(row[COL_SECONDS])
        row[COL_LIMIT] = int(row[COL_LIMIT])
        data_list.append(row)
#         print(', '.join(row))
#     csv_reader = csv.DictReader(csv_file)
#     for row in csv_reader:
#         print("Row", row['SECONDS'])
#         print("Row", row)
```

```python
def filter_date(data, filter_1, col_1, filter_2, col_2):
    new_list = []
    
    for row in data:
#         print('row[col]', row[col], filter_condition, row[col] == filter_condition)
        if row[col_1] == filter_1 and row[col_2] == filter_2:
            new_list.append(row)
    
    return new_list

def filter_value(limit_schedule, col):
    new_list = []
    
    for row in limit_schedule:
        new_list.append(row[col])
    
    return new_list
```
